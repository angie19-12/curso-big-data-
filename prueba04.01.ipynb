```scala
def medirTiempo[T](bloque: => T): (T, Long) = {
  val inicio = System.nanoTime()
  val resultado = bloque
  val fin = System.nanoTime()
  val tiempoMs = (fin - inicio) / 1000000
  (resultado, tiempoMs)
}
```

```scala
def calcularTamanoBytes(path: String): Long = {
  val p = Paths.get(path)
  if (!Files.exists(p)) 0L
  else {
    Files.walk(p)
      .filter(Files.isRegularFile(_))
      .mapToLong(Files.size(_))
      .sum()
  }
}
```

```scala
import java.io.File

def obtenerTamanioMB(ruta: String): Double = {
  val directorio = new File(ruta)
  if (directorio.exists()) {
    // Sumamos el tamaño de todos los archivos dentro de la carpeta
    val bytes = directorio.listFiles().map(_.length()).sum
    (bytes.toDouble / (1024 * 1024)) // Convertir bytes a Megabytes
  } else {
    0.0
  }
}

// Obtenemos los tamaños
val mbAvro = obtenerTamanioMB("datos_analiticos.avro")
val mbOrc = obtenerTamanioMB("datos_analiticos.orc")
val mbParquet = obtenerTamanioMB("datos_analiticos.parquet")
```

```scala
def medirTiempo(formato: String, ruta: String): Long = {
  val inicio = System.currentTimeMillis()
 
  spark.read.format(formato).load(ruta).count()
  val fin = System.currentTimeMillis()
  fin - inicio
}

val tiempoAvro = medirTiempo("avro", "datos_analiticos.avro")
val tiempoOrc = medirTiempo("orc", "datos_analiticos.orc")
val tiempoParquet = medirTiempo("parquet", "datos_analiticos.parquet")
```
