# 🧪 Pregunta 8: Análisis de temperaturas semanales

Una estación meteorológica registra la **temperatura media semanal** (en grados Celsius) durante un trimestre (12 semanas):

```scala
val temperaturas = List(
  12.5, 13.1, 14.8, 15.2,
  16.0, 18.3, 19.1, 20.0,
  18.7, 17.2, 15.9, 14.4
)

val semanas = List(
  "Sem 1", "Sem 2", "Sem 3", "Sem 4",
  "Sem 5", "Sem 6", "Sem 7", "Sem 8",
  "Sem 9", "Sem 10", "Sem 11", "Sem 12"
)
```

Tareas:

- Usa `zip` para emparejar cada semana con su temperatura.
- Imprime las **3 semanas con mayor temperatura** (usa `sortBy` y `take`).
- Calcula e imprime:
  - la **temperatura media del trimestre**
  - la **temperatura máxima registrada**
- Imprime cuántas semanas estuvieron **por encima de la media**.

### 📤 Salida de referencia (formato orientativo)

```scala
Top 3 semanas más cálidas:
Sem 8 -> 20.0 ºC
Sem 7 -> 19.1 ºC
Sem 9 -> 18.7 ºC

Temperatura media: 16.63 ºC
Temperatura máxima: 20.0 ºC
Semanas por encima de la media: 6
```


````scala

val temperaturas = List(
  12.5, 13.1, 14.8, 15.2,
  16.0, 18.3, 19.1, 20.0,
  18.7, 17.2, 15.9, 14.4
)

val semanas = List(
  "Sem 1", "Sem 2", "Sem 3", "Sem 4",
  "Sem 5", "Sem 6", "Sem 7", "Sem 8",
  "Sem 9", "Sem 10", "Sem 11", "Sem 12"
)


val registro = semanas.zip(temperaturas)
val top3 = registro.sortBy(_._2).reverse.take(3)
val promedio = temperaturas.sum / temperaturas.length
val maxima = temperaturas.max
val sobreMedia = temperaturas.count(_ > promedio)

// --- Impresiones ---
println(s"Top 3 semanas calurosas: $top3")
println(f"Temperatura media: $promedio%.2fºC")
println(s"Temperatura máxima: ${maxima}ºC")
println(s"Semanas por encima de la media: $sobreMedia")

````

````scala

val reservas: List[(String, String, Double)] = List(
  ("101", "Individual", 85.0),
  ("102", "Doble", 120.0),
  ("103", "Suite", 250.0),
  ("104", "Doble", 130.0),
  ("105", "Individual", 90.0),
  ("106", "Suite", 275.0),
  ("107", "Doble", 125.0),
  ("108", "Individual", 80.0)
)
val totalDeReservas : Int = reservas.size
println(s"Total de reservas : $totalDeReservas")

val tiposUnicos = reservas.map(_._2).toSet
println(s"Tipos de habitación : $tiposUnicos")
val porTipo = reservas.groupBy(_._2)

val precioMedioPorTipo = porTipo.map { case (tipo, listaReservas) =>
  val precios = listaReservas.map(_._3)
  val promedio = precios.sum / precios.size
  (tipo, promedio)
}

println("--- Precio Medio por Tipo ---")
precioMedioPorTipo.foreach { case (tipo, precio) =>
  println(f"$tipo%-12s : $$ $precio%.2f")
}
println("\n=== Habitación más cara por categoría ===")


val masCaro = reservas
  .groupBy(_._2)
  .view
  .mapValues(grupo => grupo.maxBy(_._3)._1)
  .toMap
masCaro.foreach { case (cat, habt) => 
  println(s"Categoría: $cat - Habitación: $habt") 
}
val importes = reservas.map(_._3)
val (minVenta, maxVenta, mediaVenta) = (importes.min, importes.max, importes.sum / importes.length)
println(f"\n=== Estadísticas Globales ===")
println(f" Venta mínima  : $minVenta%8.2f")
println(f" Venta máxima  : $maxVenta%8.2f")
println(f" Ticket medio  : $mediaVenta%8.2f")

println(f" Facturación   : ${importes.sum}%8.2f")
````
