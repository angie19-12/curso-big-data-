```scala
// Primero necesitamos un DataFrame "plano" (sin arrays) para ver el efecto
val compras = Seq(
  (1, "Ana García",   "electrónica"),
  (1, "Ana García",   "hogar"),
  (1, "Ana García",   "electrónica"),
  (2, "Pedro López",  "alimentación"),
  (2, "Pedro López",  "hogar"),
  (2, "Pedro López",  "alimentación"),
  (3, "María Ruiz",   "electrónica"),
  (3, "María Ruiz",   "deportes")
).toDF("id", "nombre", "categoria")

val resumenPorCliente = compras.groupBy("id", "nombre").agg(
  collect_list(col("categoria")).as("historial_completo"),   // CON duplicados
  collect_set(col("categoria")).as("categorias_distintas")   // SIN duplicados
).orderBy("id")

resumenPorCliente.show(truncate = false)
```

```scala
// Paso 1: explotar el array original
val explodido = clientes
  .filter(size(col("categorias")) > 0)   // excluir arrays vacíos
  .select(col("id"), col("nombre"), explode(col("categorias")).as("categoria"))

// Paso 2: ranking global de categorías más compradas
val rankingCategorias = explodido
  .groupBy("categoria")
  .count()
  .orderBy(col("count").desc)

println("🏆 Ranking de categorías:")
rankingCategorias.show(truncate = false)

// Paso 3: resumen por cliente con categorías únicas y total de compras
val resumenClientes = explodido
  .groupBy("id", "nombre")
  .agg(
    count("categoria").as("total_compras"),
    collect_set("categoria").as("especialidades")
  )
  .withColumn("num_especialidades", size(col("especialidades")))
  .orderBy("id")

println("📋 Resumen por cliente:")
resumenClientes.show(truncate = false)
```

```scala
// Usamos Option[Double] para que Spark entienda que puede haber nulos
val clientes = Seq(
  (1, "Ana García", Some(1250.0)),
  (2, "Pedro López", Some(320.5)),
  (3, "María Ruiz", Some(85.0)),
  (4, "Carlos Sanz", Some(780.0)),
  (5, "Laura Vega", Some(2400.0)),
  (6, "Jorge Martín", None) // Esto representa el valor nulo de forma segura
).toDF("id", "nombre", "total_gasto")

clientes.show(truncate = false)
```

```scala
import org.apache.spark.sql.functions.udf


val clasificarGasto: java.lang.Double => String = (gasto) => {
  if      (gasto == null)   "Desconocido"
  else if (gasto >= 1000.0) "Premium"
  else if (gasto >= 300.0)  "Estándar"
  else                       "Básico"
}


val clasificarGastoUDF = udf(clasificarGasto)

val clientesConSegmento = clientes
  .withColumn("segmento", clasificarGastoUDF(col("total_gasto")))

clientesConSegmento.show(truncate = false)
```

```scala
// UDF INCORRECTA: usa Double primitivo, no gestiona null
val clasificarGastoRoto: Double => String = (gasto) => {
  if      (gasto >= 1000.0) "Premium"
  else if (gasto >= 300.0)  "Estándar"
  else                       "Básico"
}

val clasificarGastoRotaUDF = udf(clasificarGastoRoto)

// Esto lanza NullPointerException al llegar a la fila con null
val clientesError = clientes
  .withColumn("segmento_roto", clasificarGastoRotaUDF(col("total_gasto")))

clientesError.show()   // ← falla aquí en tiempo de ejecución
```

```scala

spark.udf.register("clasificar_gasto", clasificarGasto)

// Crear la vista temporal para poder referenciarla en SQL
clientes.createOrReplaceTempView("clientes")


val resultadoSQL = spark.sql("""
  SELECT
    id,
    nombre,
    total_gasto,
    clasificar_gasto(total_gasto) AS segmento
  FROM clientes
  ORDER BY total_gasto DESC
""")

resultadoSQL.show(truncate = false)
```

```scala
val clientesConSegmentoNativo = clientes
  .withColumn("segmento",
    when(col("total_gasto") >= 1000.0, "Premium")
    .when(col("total_gasto") >= 300.0,  "Estándar")
    .otherwise("Básico")
  )

clientesConSegmentoNativo.show(truncate = false)
```

¿Qué diferencia hay entre explode y explode_outer?El explode si la lista esta vacía la elimina  y con explode_outer si la lista esta vacía la mantiene en la columnna.
¿Por qué se recomienda when/otherwise en lugar de una UDF para lógica condicional?Prque los UDF en Spark es mucho más lento y consume mas memoria (se usa UDF si la lógica es tan compleja que no s epuede hacer con los códigos de spark)
¿Por qué usamos java.lang.Double en lugar de Double en la UDF?Para que los códigos sean seguros y no ecxplote cuando una fila salga delante
¿Qué diferencia hay entre collect_list y collect_set? el collect list guarda todos los elementos en orden (hasta los duplicados ) y el collect_set guarda los unicos elementos(elimina los duplicados)y no se ponen en ningun orden .

```scala
import org.apache.spark.sql.functions._
import org.apache.spark.sql.types.DateType // Importamos el tipo específico
   
val pacientes = Seq(
  (" juan perez ", "JUAN@GMAIL.COM", "600-123-456", "1990-05-15", "2023-01-01"),
  ("MARIA lopez", " Maria.Lopez@Hotmail.com ", " +34 912 345 678 ", "1985-10-20", "2023-02-15")
).toDF("nombre", "email", "telefono", "fecha_nacimiento", "fecha_alta")


val pacientesLimpios = pacientes
  .withColumn("nombre", initcap(trim($"nombre"))) // Corregido paréntesis
  .withColumn("email", lower(trim($"email")))
  .withColumn("telefono", regexp_replace($"telefono", "[^0-9]", "")) // Corregido 'regexp'
  .withColumn("fecha_nacimiento", $"fecha_nacimiento".cast(DateType))
  .withColumn("fecha_alta", $"fecha_alta".cast(DateType))

pacientesLimpios.show(truncate = false)
pacientesLimpios.printSchema()
```

```scala
import org.apache.spark.sql.functions._


val pacientesEnriquecidos = pacientesLimpios

  .withColumn("id", monotonically_increasing_id()) 
  .withColumn("edad", (months_between(current_date(), col("fecha_nacimiento")) / 12).cast("int"))
  .withColumn("anios_como_paciente", (months_between(current_date(), col("fecha_alta")) / 12).cast("int"))
  .withColumn("fecha_alta_formato_es", date_format(col("fecha_alta"), "dd/MM/yyyy"))
  .withColumn("decada_nacimiento", (floor(year(col("fecha_nacimiento")) / 10) * 10).cast("int"))


pacientesEnriquecidos.select(
  col("id"), 
  col("nombre"), 
  col("edad"), 
  col("anios_como_paciente"), 
  col("fecha_alta_formato_es"), 
  col("decada_nacimiento")
).show(truncate = false)
```

```scala
import org.apache.spark.sql.functions._


val pacientesImagen = Seq(
  (1, "Ana García Torres", 2),
  (2, "Pedro López", 5),
  (3, "María Ruiz", 1),
  (4, "Carlos Sanz", 4),
  (5, "Laura Vega", 3),
  (6, "Jorge Martínez", 7),
  (7, "Sofía Ramos", 2),
  (8, "Antonio Vidal", 9)
).toDF("id", "nombre", "num_visitas")
val resultadoFinal = pacientesImagen
  .withColumn("perfil_visitas", 
    when(col("num_visitas") >= 7, "Frecuente")
    .when(col("num_visitas") >= 4, "Habitual")
    .when(col("num_visitas") >= 2, "Ocasional")
    .otherwise("Nuevo")
  )
resultadoFinal.select("id", "nombre", "num_visitas", "perfil_visitas")
  .show(truncate = false)
```

```scala
import org.apache.spark.sql.functions._


val tratamientosRaw = Seq(
  (1, "Pediatría;Fisioterapia;Pediatría"),
  (2, "Cardiología;Neurología"),
  (3, ""), 
  (4, "Odontología;Odontología;Odontología")
).toDF("id", "tratamientos_texto")


val tratamientosArray = tratamientosRaw
  .withColumn("tratamientos_array", split(col("tratamientos_texto"), ";"))


val tratamientosLimpio = tratamientosArray
  .filter(size(col("tratamientos_array")) > 0 && col("tratamientos_array")(0) =!= "")


val resultadoFinal = tratamientosLimpio
  .withColumn("num_tratamientos_totales", size(col("tratamientos_array")))
  .withColumn("tratamientos_unicos", array_distinct(col("tratamientos_array")))
  .withColumn("num_especialidades", size(col("tratamientos_unicos")))

// Mostrar resultado
resultadoFinal.show(truncate = false)
```
