```scala
// 1. Cargamos la librería
import $ivy.`org.apache.spark::spark-sql:3.4.1`

import org.apache.spark.sql.SparkSession

// 2. Creamos la sesión (Mucho más simple ahora)
val spark = SparkSession.builder()
  .appName("AgroData_Analisis")
  .master("local[*]")
  .getOrCreate()

// 3. Configuración de logs
spark.sparkContext.setLogLevel("ERROR")

// 4. Mensaje de éxito
println(s"✅ ¡POR FIN! Spark ${spark.version} funcionando correctamente con Java 11")
```

```scala
import spark.implicits._

val datosParcelas = Seq(
  ("P001","Sevilla","Carmona",12.5,"Naranja",2015,true,28000),
  ("P002","Huelva","Lepe",8.3,"Fresa",2018,true,45000),
  ("P003","Almería","Níjar",25.0,"Tomate",2012,true,85000),
  ("P004","Sevilla","Écija",6.7,"Aceituna",2009,false,3200),
  ("P005","Murcia","Totana",15.2,"Limón",2016,true,22000),
  ("P006","Almería","El Ejido",30.1,"Pimiento",2014,true,62000),
  ("P007","Huelva","Moguer",9.8,"Fresa",2020,true,41000),
  ("P008","Murcia","Lorca",18.4,"Melocotón",2011,false,9500),
  ("P009","Sevilla","Utrera",22.0,"Naranja",2013,true,31000),
  ("P010","Almería","Vícar",11.6,"Pepino",2019,true,74000),
  ("P011","Murcia","Alhama",7.9,"Limón",2017,true,20500),
  ("P012","Huelva","Cartaya",14.3,"Fresa",2015,true,43000),
  ("P013","Sevilla","Marchena",5.2,"Aceituna",2010,false,2900),
  ("P014","Almería","Roquetas",28.7,"Tomate",2011,true,88000),
  ("P015","Murcia","Mazarrón",16.5,"Pimiento",2018,true,58000)
) // <--- ESTE ES EL PARÉNTESIS QUE FALTABA

val dfParcelas = datosParcelas.toDF("Id_parcela", "provincia", "municipio", "superficie_ha", "cultivo", "año_alta", "en_produccion", "rendimiento_kg_ha")

println("--- Muestra de las primeras 5 parcelas ---")
dfParcelas.show(5)

println("--- Schema ---")
dfParcelas.printSchema()

println(s"Total parcelas : ${dfParcelas.count()}")

println("--- Columnas ---")
println(dfParcelas.columns.mkString(", "))
```

```scala
import org.apache.spark.sql.types._

val schemaParcelas = StructType(Array(
  StructField("id_parcela", StringType, false),
  StructField("provincia", StringType, true),
  StructField("municipio", StringType, true),
  StructField("superficie_ha", DoubleType, true),
  StructField("cultivo", StringType, true),
  StructField("año_alta", IntegerType, true),
  StructField("en_produccion", BooleanType, true),
  StructField("rendimiento_kg_ha", IntegerType, true)
))
```

```scala
import org.apache.spark.sql.Row

val dfConSchema = spark.createDataFrame(
  spark.sparkContext.parallelize(datosParcelas.map(Row.fromTuple)),
  schemaParcelas
)

println("--- Nuevo Schema Manual ---")
dfConSchema.printSchema()

println("--- Comparativa de tipos (columna - tipo) ---")
dfConSchema.dtypes.foreach { case (columna, tipo) =>
  // Usamos 'f' antes de las comillas para que el %-20s funcione
  println(f"$columna%-20s -> $tipo") 
}
```

```scala
println("RESUMEN ESTADÏSTICO:Superficie y Rendimiento")
dfConSchema.describe("superficie_ha","rendimiento_kg_ha").show()

println("ANÄLISIS DE ANTIGÜEDAD :Año de Alta")
dfConSchema.describe("año_alta").show()
```

```scala
import spark.implicits._

val datosProductos = Seq(
  ("Naranja", 1.20, true),
  ("Fresa", 3.50, true),
  ("Tomate", 2.10, false),
  ("Aceituna", 0.85, false),
  ("Limón", 1.45, true),
  ("Pimiento", 2.30, true),
  ("Melocotón", 1.90, false),
  ("Pepino", 1.15, true)
)

val dfProductos = datosProductos.toDF("nombre_producto", "precio_mercado_euro_kg", "certificacion_eco")



println("--- Listado Completo de productos ---")
dfProductos.show(truncate = false)

println("--- Estructura de productos ---")
dfProductos.printSchema()


```

```scala
import org.apache.spark.sql.types._
import org.apache.spark.sql.Row

val schemaProductosManual = StructType(Array(
  StructField("codigo", StringType, false),
  StructField("nombre_comercial", StringType, true),
  StructField("precio_mercado_euro_kg", DoubleType, false),
  StructField("certificacion_eco", BooleanType, true),
  StructField("stock_disponible", IntegerType, true) // Corregido: IntegerType
))

val dfProductosManual = spark.createDataFrame(
  spark.sparkContext.parallelize(datosProductos.map(Row.fromTuple)), // Corregido: parallelize
  schemaProductosManual
)

println("--- Tipos con Esquema Manual (Comercial) ---")
dfProductosManual.dtypes.foreach { case (col, tipo) =>
  println(f"$col%-25s -> $tipo")
}
```
